```python
import os
import numpy as np 
from sklearn.preprocessing import MinMaxScaler
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
flux_arr=[]
for file in Path("dataset").iterdir():
    a = np.load(file)
    flux = a['flux']
    if np.isnan(flux).any():
        median_val = np.nanmedian(flux)
        flux = np.where(np.isnan(flux), median_val, flux)

    flux_arr.extend(flux)
    print(flux)

flux_arr = np.array(flux_arr)

plt.figure(figsize=[8,5])
plt.hist(flux_arr, bins=10, color="skyblue", edgecolor="black")
plt.xlabel("Flux")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# data already normalized
```
